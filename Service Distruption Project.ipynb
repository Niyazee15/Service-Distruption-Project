```python
# Converting categoricals using get dummies
towers = pd.get_dummies(towers, drop_first = True)
towers.head()
```

```python
#Groupby 'id'

towers = towers.groupby(['id']).sum()
```

```python
towers.shape
```

```python
# reseting id to a column which was listed as a index
towers.reset_index(inplace=True)
```

```python
# need to add the original data set which had fault severity to this column
newdata = pd.merge(tower,towers)
```

```python
newdata.head()
```

```python
#newdata.drop('id', axis=1, inplace=True)
```

```python
newdata.drop('location', axis=1, inplace=True)
```

```python
newdata.head()
```

```python
x = newdata.drop('fault_severity', axis=1)
```

```python
y = newdata.fault_severity
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=22)
```

```python
# use Gradient Boosting 
from sklearn.ensemble import GradientBoostingClassifier
```

```python
towert = GradientBoostingClassifier()

towert.fit(x_train, y_train)
#towert.score(x_test, y_test)
```

```python
y_pred = towert.predict(x_test)
y_pred_proba_log = towert.predict_proba(x_test)
```

```python
y_pred
```

```python
y_pred_proba_log
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression

LogicR = LogisticRegression()

LogicR.fit(x_train, y_train)

LogicR.score(x_test, y_test)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier

RandF = RandomForestClassifier()
RandF.fit(x_train,y_train)
RandF.score(x_test, y_test)
```

```python
result = pd.DataFrame({
    "id": x_test.id,
    "Prediction fault_severity": y_pred,
    "Prediction_probability_0" : y_pred_proba_log[:,0],
    "Prediction_probability_1" : y_pred_proba_log[:,1],
    "Prediction_probability_2" : y_pred_proba_log[:,2],
},columns=['id','Prediction fault_severity','Prediction_probability_0','Prediction_probability_1','Prediction_probability_2'])

```
